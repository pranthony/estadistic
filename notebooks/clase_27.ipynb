```python
# Reimportar libreras y recrear el grfico

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Crear valores para el eje X
x = np.linspace(-4, 4, 1000)

# Distribucin normal estndar
normal_dist = norm.pdf(x, 0, 1)

# Crear la figura
plt.figure(figsize=(10, 6))

# Graficar la curva de distribucin normal
plt.plot(x, normal_dist, label='Distribucin Normal', color='blue')

# Sombrear las reas para 1, 2, y 3
plt.fill_between(x, normal_dist, where=(x >= -1) & (x <= 1), color='green', alpha=0.3, label='68% de los datos (1)')
plt.fill_between(x, normal_dist, where=(x >= -2) & (x <= 2), color='yellow', alpha=0.3, label='95% de los datos (2)')
plt.fill_between(x, normal_dist, where=(x >= -3) & (x <= 3), color='red', alpha=0.2, label='99.7% de los datos (3)')

# Aadir lneas verticales en 1, 2, y 3
plt.axvline(0, color='black', linestyle='-', label='Media (x)')
plt.axvline(1, color='green', linestyle='--')
plt.axvline(-1, color='green', linestyle='--')
plt.axvline(2, color='yellow', linestyle='--')
plt.axvline(-2, color='yellow', linestyle='--')
plt.axvline(3, color='red', linestyle='--')
plt.axvline(-3, color='red', linestyle='--')

# Aadir ttulo y etiquetas
plt.title('Ilustracin de la Desviacin Estndar en una Distribucin Normal')
plt.xlabel('Observaciones (X)')
plt.ylabel('Densidad de Probabilidad')

# Aadir leyenda
plt.legend()

# Mostrar el grfico
plt.grid(True)
plt.show()
```
