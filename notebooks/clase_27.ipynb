```python
# Reimportar librerías y recrear el gráfico

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Crear valores para el eje X
x = np.linspace(-4, 4, 1000)

# Distribución normal estándar
normal_dist = norm.pdf(x, 0, 1)

# Crear la figura
plt.figure(figsize=(10, 6))

# Graficar la curva de distribución normal
plt.plot(x, normal_dist, label='Distribución Normal', color='blue')

# Sombrear las áreas para ±1σ, ±2σ, y ±3σ
plt.fill_between(x, normal_dist, where=(x >= -1) & (x <= 1), color='green', alpha=0.3, label='68% de los datos (±1σ)')
plt.fill_between(x, normal_dist, where=(x >= -2) & (x <= 2), color='yellow', alpha=0.3, label='95% de los datos (±2σ)')
plt.fill_between(x, normal_dist, where=(x >= -3) & (x <= 3), color='red', alpha=0.2, label='99.7% de los datos (±3σ)')

# Añadir líneas verticales en ±1σ, ±2σ, y ±3σ
plt.axvline(0, color='black', linestyle='-', label='Media (x)')
plt.axvline(1, color='green', linestyle='--')
plt.axvline(-1, color='green', linestyle='--')
plt.axvline(2, color='yellow', linestyle='--')
plt.axvline(-2, color='yellow', linestyle='--')
plt.axvline(3, color='red', linestyle='--')
plt.axvline(-3, color='red', linestyle='--')

# Añadir título y etiquetas
plt.title('Ilustración de la Desviación Estándar en una Distribución Normal')
plt.xlabel('Observaciones (X)')
plt.ylabel('Densidad de Probabilidad')

# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
